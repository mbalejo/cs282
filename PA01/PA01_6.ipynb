```python
# PA01: Image Enhancement
# Written by: Marwin B. Alejo 2020-20221
# February 23, 2022
#
# Item 6 Requirement: 
# Unsharp masking of building.tif
#
# ATTENTION: Proper reading of TIF file required in this task to prevent data loss. 

import cv2 as cv
from scipy.ndimage.filters import median_filter
from google.colab.patches import cv2_imshow
from google.colab import drive
drive.mount('/content/drive/', force_remount=True)

image = cv.imread('/content/drive/MyDrive/Colab_Notebooks/building.tif', cv.IMREAD_COLOR) # this will read the image as an RGB array
blurred = median_filter(image, 1) # perform median blurring
laplacedImg = cv.Laplacian(blurred, cv.CV_64F) # edge detector on blurred image
maskImg = image - 0.7*laplacedImg # unsharp masking
print("Before Unsharp Masking")
cv2_imshow(image)
print("\n\nAfter Unsharp Masking")
cv2_imshow(maskImg)
```
